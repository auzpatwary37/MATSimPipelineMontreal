```python
def compute_age_category(age):
    if age > 80:
        result = 16
    else:
        result = age // 5 + 1
    return result
```

## Driving license

```python
saaq_data= pd.read_excel('SAAQ_AD_donnees_all_2016.xlsx', sheet_name="2016_csv")
```

```python
#pip install pyogrio
```

```python
geographical_filter = gpd.read_file('spatial/local_zoning.shp')
```

```python
#saaq_data = pd.read_csv("../data/quebec_data/SAAQ_driving_license_car_ownership.csv", sep=";")
#saaq_data = saaq
#saaq_data = saaq_data[:-1].copy()  # Drop the last record with no specific geography
saaq_data1 = saaq_data[:-1]
saaq_data1['adidu'] = saaq_data1['adidu'].astype(int)
saaq_data1
```

```python
saaq_data = saaq_data[saaq_data.adidu.isin(geographical_filter.DAUID)]
```

```python
saaq_data.drop(columns=['auto_elec','auto_conv', 'camleg_elec', 'camleg_conv', 'nonprec_elec','nonprec_conv', 'auto_tot', 'camleg_tot', 'nonprec_tot', 'elec_tot','conv_tot',
                        'commentaires'], inplace=True)
```

```python
saaq_data.info()
```

```python
saaq_data.rename(columns={"permis_h":"license_sex_1", "permis_f": "license_sex_2", "permis_tot": "license", "nbadr_1permis": "hh_driving_license_1", "nbadr_2permis": "hh_driving_license_2",
                          "nbadr_3permis": "hh_driving_license_3", "nbadr_4pluspermis": "hh_driving_license_4", "tot_veh": "hh_car", "nbadr_1pau": "hh_car_1", "nbadr_2pau": "hh_car_2",
                          "nbadr_3pau": "hh_car_3", "nbadr_4pluspau": "hh_car_4"}, inplace=True)
```

```python
saaq_data.fillna(np.random.randint(1,6), inplace=True)
```

```python
saaq_data
```

```python
saaq_data = saaq_data.merge(geographical_filter, left_on="adidu", right_on="DAUID")
```

```python
saaq_data[['DAUID', 'CTUID', 'license', 'license_sex_1', 'license_sex_2']].to_csv("output/person_license_sex.csv", index=False)
```
